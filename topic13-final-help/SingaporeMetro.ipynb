```python
class Station:
    def __init__(self, code, name, line, serial):
        self.code = code
        self.name = name
        self.line = line
        self.serial = serial
        self.connections = {}
    def connect(self, code, duration):
        if code in self.connections:
            return
        self.connections[code] = duration
    def __str__(self):
        s = self.name + ", " + self.code
        for con in self.connections:
            s += " [" + con + ": " + str(self.connections[con]) + "] "
        return s
            
```

```python
stations = {}
```

```python
for line in df_stations.values:
    stations[line[1]] = Station(line[1], line[0], line[3], line[4])
```

```python
import random
```

```python
for code in stations:
    station = stations[code]
    next_st = station.line + str(station.serial+1)
    prev_st = station.line + str(station.serial-1)
    if next_st in stations:
        duration = random.randint(3, 7)
        station.connect(next_st, duration)
        stations[next_st].connect(station.code, duration)
    if prev_st in stations:
        duration = random.randint(3, 7)
        station.connect(prev_st, duration)
        stations[prev_st].connect(station.code, duration)
    # Find all stations with same name
    for code2 in stations:
        station2 = stations[code2]
        if station.name == station2.name and station.code != station2.code:
            station.connect(station2.code, 5)
            station2.connect(station.code, 5)
```
